```python
### 3. Read dogs images from disk into numpy array using opencv
```

```python
# You need to finish the dogs_images_dict

# Include your solution below

dogs_images_dict = {
    'beagle': list(data_dir.glob('beagle/*.jpg')),
    'bedlington_terrier': list(data_dir.glob('bedlington_terrier/*.jpg')),
    'pug': list(data_dir.glob('pug/*.jpg')),
}
# Display the first few image paths for each breed
print("Beagle Images:", dogs_images_dict['beagle'][:5])
print("Bedlington Terrier Images:", dogs_images_dict['bedlington_terrier'][:5])
print("Pug Images:", dogs_images_dict['pug'][:5])

```

```python
dogs_labels_dict = {
    'beagle': 0,
    'bedlington_terrier': 1,
    'pug': 2,
}
```

```python
dogs_images_dict['pug'][:5]
```

```python
str(dogs_images_dict['pug'][0])
```

```python
img = cv2.imread(str(dogs_images_dict['pug'][0]))
img.shape
```

```python
cv2.resize(img,(224,224)).shape
```

```python
X, y = [], []

for dog_name, images in dogs_images_dict.items():
    for image in images:
        img = cv2.imread(str(image))
        resized_img = cv2.resize(img,(224,224))
        X.append(resized_img)
        y.append(dogs_labels_dict[dog_name])
```

```python
X = np.array(X)
y = np.array(y)
```

```python
## 4. Train test split
```

```python
from sklearn.model_selection import train_test_split
# Include your solution below

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1)

X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.25, random_state=1)

#In this way, train, val, test set will be 60%, 20%, 20% of the dataset respectively.
```

```python
# 4.1 Representing Labels using one-hot-ecoding
```

```python
k = np.unique(y).size
Y_oh_train = to_categorical(y_train, k)
Y_oh_val = to_categorical(y_val, k)
Y_oh_test = to_categorical(y_test, k)
# Displaying the 5 first elemnts
print('Y_train[:5]:')
print(y_train[:5])
print('\nY_oh_train[:5]=')
print(Y_oh_train[:5])
```

```python
## 5. Preprocessing: scale images
```

```python
# 7. Now take pre-trained model and retrain it using flowers images
```

```python
feature_extractor_model = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"

pretrained_model_without_top_layer = hub.KerasLayer(
    feature_extractor_model, input_shape=(224, 224, 3), trainable=False)
```

```python
num_of_dogclasses = 3

model = tf.keras.models.Sequential([
  pretrained_model_without_top_layer,
  tf.keras.layers.Dense(num_of_dogclasses)
])

model.summary()


```

```python
#categorical_crossentropy you use one hot encoding,

#and if you use sparse_categorical_crossentropy you encode as normal integers.
```

```python
model.compile(
  optimizer="adam",
  loss=tf.keras.losses.CategoricalCrossentropy(),
  metrics=['accuracy'])

history = model.fit(X_train_scaled, Y_oh_train, validation_data=(X_val_scaled,Y_oh_val),batch_size= 64, epochs= 5)
```

```python
model.evaluate(X_test_scaled,Y_oh_test)
```

```python
model.compile(
  optimizer="adam",
  loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
  metrics=['accuracy'])

history = model.fit(X_train_scaled, y_train, validation_data=(X_val_scaled,y_val),batch_size= 64, epochs= 5)
```

```python
model.evaluate(X_test_scaled,y_test)
```
